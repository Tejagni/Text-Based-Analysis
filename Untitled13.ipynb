## pipeline

```python
pipeline = [str.lower, tokenize]
def prepare(text, pipeline):
    tokens = text
    for fun in pipeline:
        tokens = fun(tokens)
    return tokens
```

## running the pipeline

```python
df['tokens'] = df['Review_Text'].apply(prepare, pipeline=pipeline)
```

```python
from collections import Counter
tokens = tokenize("I love apple and apple is a fruit")
testcount= Counter(tokens)
print(testcount)
```

```python
# Counting the Word Repetations
counter = Counter()
df['tokens'].map(counter.update)
print(counter.most_common(5))
```

```python
# Using word frequency in it’s own dataframe for analysis
freq_df = pd.DataFrame.from_dict(counter, orient='index', columns=['freq']) 
freq_df = freq_df.query('freq >= 1') # Use it to Prevent Worlds that are infrequent
freq_df.index.name = 'token'
```

```python
# Let,s sort Based on Frequency and display
freq_df = freq_df.sort_values('freq', ascending=False)
freq_df.head(10)
```

## 4-regex

```python
import re

text = "Contact us at +1 (555) 123-4567"
phone_pattern = r'\+?[0-9]+[\s-]?[\(]?[0-9]{3}[\)]?[\s-]?[0-9]{3}[\s-]?[0-9]{4}'

match = re.search(phone_pattern, text)
if match:
    print("Phone number found:", match.group(0))
else:
    print("No phone number found.")
```

```python
import re

text = "Open the file located at /path/to/file.txt"
file_path_pattern = r'\/([^\/]+\/)*[^\/]+\.\w+'

match = re.search(file_path_pattern, text)
if match:
    print("File path found:", match.group(0))
else:
    print("No file path found.")
```

```python
import re

text = "The event will take place on 2024-04-27"
date_pattern = r'\b\d{4}-\d{2}-\d{2}\b'

match = re.search(date_pattern, text)
if match:
    print("Date found:", match.group(0))
else:
    print("No date found.")
```

```python
import re

text = "Please send the report to example@email.com"
email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'

match = re.search(email_pattern, text)
if match:
    print("Email address found:", match.group(0))
else:
    print("No email address found.")

    
```

```python
import re

txt = "Chichili Tejagni Reddy"

#Find all characters alphabetically between "s"-"y" and "H"-"Y":

x = re.findall("[s-yH-Y]", txt)
print(x)
```

```python
import re

text = "The tutorial is available at http://www.exampletutorial.com"
url_pattern = r'(https?://[^\s]+)'

match = re.search(url_pattern, text)
if match:
    print("URL found:", match.group(0))
else:
    print("No URL found.")
```

```python
#Clean the text
#Create a function to clean the Contents
def cleanTxt(Review_Text):
    CONTENT = re.sub(r'@[A-Za-z0-9]+', '', CONTENT) #Remove @mentions replace with blank
    CONTENT = re.sub(r'#', '', CONTENT) #Remove the '#' symbol, replace with blank
    CONTENT = re.sub(r'RT[\s]+', '', CONTENT) #Removing RT, replace with blank
    CONTENT = re.sub(r'https?:\/\/\S+', '', CONTENT) #Remove the hyperlinks
    CONTENT= re.sub(r':', '', CONTENT) # Remove :
    
    return Review_Text
#Cleaning the CONTENT
df["CONTENT"]=df["CONTENT"].apply(cleanTxt)

#Show the clean CONTENT
df.head()
```
