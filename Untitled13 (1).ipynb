## 4-regex

```python
import re

text = "Contact us at +1 (555) 123-4567"
phone_pattern = r'\+?[0-9]+[\s-]?[\(]?[0-9]{3}[\)]?[\s-]?[0-9]{3}[\s-]?[0-9]{4}'

match = re.search(phone_pattern, text)
if match:
    print("Phone number found:", match.group(0))
else:
    print("No phone number found.")
```

```python
import re

text = "Open the file located at /path/to/file.txt"
file_path_pattern = r'\/([^\/]+\/)*[^\/]+\.\w+'

match = re.search(file_path_pattern, text)
if match:
    print("File path found:", match.group(0))
else:
    print("No file path found.")
```

```python
import re

text = "The event will take place on 2024-04-27"
date_pattern = r'\b\d{4}-\d{2}-\d{2}\b'

match = re.search(date_pattern, text)
if match:
    print("Date found:", match.group(0))
else:
    print("No date found.")
```

```python
import re

text = "Please send the report to example@email.com"
email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'

match = re.search(email_pattern, text)
if match:
    print("Email address found:", match.group(0))
else:
    print("No email address found.")

    
```

```python
#Clean the text
#Create a function to clean the Contents
def cleanTxt(Review_Text):
    CONTENT = re.sub(r'@[A-Za-z0-9]+', '', Review_Text) 
    CONTENT = re.sub(r'#', '', Review_Text) 
    CONTENT = re.sub(r'RT[\s]+', '', Review_Text) 
    CONTENT = re.sub(r'https?:\/\/\S+', '', Review_Text)
    CONTENT= re.sub(r':', '', Review_Text) 
    
    return Review_Text

df["https://app.reviewnb.com/Tejagni/Text-Based-Analysis/pull/9/"]=df["Review_Text"].apply(cleanTxt)

df.head()
```

## removing emoji & Unicode from the review text Column

```python
def remove_emoji(string):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)

#Cleaning the text
df["Review_Text"]=df["Review_Text"].apply(remove_emoji)

#Show the clean text
df.head()
```

```python
from textblob import TextBlob ## Importing TextBlob
from wordcloud import WordCloud, STOPWORDS  
```

```python
#Create a function to get the subjectivity
def getSubjectivity(Review_Text):
    return TextBlob(Review_Text).sentiment.subjectivity

#Create a function to get Polarity
def getPolarity(Review_Text):
    return TextBlob(Review_Text).sentiment.polarity

#Now we create a new column for what we just did and add it to the youtube dataframe
df['Subjectivity'] = df['Review_Text'].apply(getSubjectivity)
df['Polarity'] = df['Review_Text'].apply(getPolarity)

#Now display data
df.head()
```

```python
## grouping the polarities
```

```python
def getInsight(score):
    if score < 0:
        return "Negative"
    elif score == 0:
        return "Neutral"
    else:
        return "Positive"
    
df["Insight"] =df["Polarity"].apply(getInsight)

df.head(50)
```
