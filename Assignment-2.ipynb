## regex operations

## To pull out words

```python
import regex as re
def tokenize(text):
    return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', text)
x=tokenize('Hello world')
print(x)
```

## Function to tokenize

```python
testtext = "Let's defeat SARS-CoV-2 together in 2020! www.cdc.gov"
testtokens = tokenize(testtext)
print(testtokens)
```

## stopwords

```python
import nltk 
from nltk.tokenize import word_tokenize
nltk.download('punkt') 
nltk.download('stopwords')
stopwords = set(nltk.corpus.stopwords.words('english'))  
print(stopwords)
```

## Remove stop words

```python
def remove_stop(tokens):
    return [t for t in tokens if t.lower() not in stopwords]
#Just testing it
cleantokens= remove_stop(testtokens)
print(cleantokens)
```

## Pipeline 

```python
pipeline = [str.lower, tokenize, remove_stop]
def prepare(text, pipeline):
    tokens = text
    for fun in pipeline:
        tokens = fun(tokens)
    return tokens
```

## Run the pipeline

```python
df['tokens'] = df['ticket'].apply(prepare, pipeline=pipeline)
```

## count the tokens

```python
df['num_tokens'] = df['tokens'].map(len)
df.tail()
```

## Word Frequency Analysis

```python
from collections import Counter
tokens = tokenize("I love apples, apple are my favorite fruit")
testcount= Counter(tokens)
print(testcount)
```

## update

```python
more_tokens = tokenize("I like apples and bananas")
testcount.update(more_tokens)
print(testcount)
```

## Counting tokens

```python
counter = Counter()
df['tokens'].map(counter.update)
print(counter.most_common(5))
```

## Analysis

```python
freq_df = pd.DataFrame.from_dict(counter,orient='index', columns=['freq'])
freq_df = freq_df.query('freq >= 1')
freq_df.index.name = 'token'
```

## Sort and display

```python
freq_df = freq_df.sort_values('freq', ascending=False)
freq_df.head(10)
```

```python
freq_df.describe()
```

## Barchart

```python
ax = freq_df.head(15).plot(kind='barh', width=0.95, figsize=(12,8))
ax.invert_yaxis()
ax.set(xlabel='Frequency', ylabel='Token', title='Top Words')
```
